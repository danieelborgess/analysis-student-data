```python
import kagglehub
import pandas as pd

# O código vai baixar os arquivos e salvar o caminho na variável 'path'
path = kagglehub.dataset_download("nawazkhan3251/student-performance-and-behavioral-analytics-dataset")

# Como sabemos que o nome do arquivo é esse, já montamos o caminho completo
caminho_csv = f"{path}/hybrid_student_performance_1200.csv"

# Carregando no Pandas!
df = pd.read_csv(caminho_csv)
```

```python
colunas_texto = ['revision_frequency',

'focus_duration',

'screen_time_non_study',

'sleep_hours',

'online_courses',

'programming_foundation',

'events_participation',

'external_resources',

'external_pressure']

for colunas in colunas_texto:
  df[colunas] = df[colunas].fillna("Não informado")

colunas_numericas = ['daily_productivity',

'energy_level',

'stress_level',

'routine_rating']

for coluna in colunas_numericas:
  df[coluna] = df[coluna].fillna(df[coluna].median())

print(df.isnull().sum())
```

### O Peso dos Projetos e Estágios no Desempenho

Ao cruzar o engajamento prático dos alunos (Projetos/Estágios) com suas notas, os dados quebram o mito de que trabalhar e estudar simultaneamente afunda o boletim:

* **O Topo é Prático:** O grupo de alunos que já atua ativamente no mercado ou em projetos ("Yes, actively working...") possui a maior concentração de notas máximas (9.5 - 10.0), atingindo 17.6%.

**Conclusão:** Ter a "mão na massa" (ou estar se planejando ativamente para isso) puxa o desempenho do aluno para cima. A experiência prática atua como um motor para o estudo teórico, e não como uma âncora.

### O Paradoxo do Estresse no Desempenho Acadêmico

Ao investigar a relação entre o Nível de Estresse (1 a 5) e as notas (CGPA), os dados revelaram um cenário não linear, quebrando a expectativa comum de que o estresse elevado sempre destrói o desempenho acadêmico:

* **Os Mais Estressados:** Surpreendentemente, os alunos com o nível máximo de estresse (Nível 5) possuem a maior taxa de notas de excelência da turma (19.1% na faixa de 9.5 - 10.0). Isso indica que a alta performance, muitas vezes, cobra um preço alto na saúde mental do estudante, ou que a própria cobrança por perfeição gera esse estresse.
* **O "Vale" do Nível Médio:** O pior cenário acadêmico não está nos extremos, mas no meio. Alunos com Nível 3 de estresse possuem a menor taxa de notas máximas (12.5%) e o maior pico de notas críticas (23.1% na faixa de 5.0 - 6.9).
* **O Equilíbrio Ideal:** O Nível 1 (baixo estresse) também apresenta ótimos resultados, com 18.5% de alunos na faixa de excelência, provando que é possível chegar ao topo sem sacrificar a saúde.

**Conclusão:** O estresse extremo não impede notas altas (pelo contrário, eles caminham juntos no topo), mas a falta de gerenciamento emocional ou a desorganização que gera um estresse "intermediário" parece ser fatal para o boletim.

```python
grafico4_pct = df.groupby("sleep_hours")["cgpa_category"].value_counts(normalize=True).reset_index(name="Percentual")

df_grafico4 = px.bar(
    grafico4_pct,
    x="sleep_hours",
    y="Percentual",
    title="Relação de Desempenho e Sono",
    color="cgpa_category",
    barmode="group",
    text_auto='.1%'
)

df_grafico4.show()
```

### O Ponto de Equilíbrio do Sono

Ao analisar a relação entre as horas de sono e as notas finais (CGPA), descobrimos que a regra de "quanto mais, melhor" não se aplica ao descanso acadêmico. O gráfico revela uma curva de eficiência muito clara:

* **A Estratégia Ideal:** O pico de excelência acadêmica não está nos extremos. Alunos que dormem entre 6 e 7 horas possuem a maior concentração de notas máximas (17.0% na faixa de 9.5 - 10.0). É o balanço perfeito entre tempo de estudo e recuperação mental.
* **O Esforço Noturno:** Alunos que dormem pouco (4-5 horas) superam aqueles que dormem muito. Eles possuem 14.5% de alunos na faixa de excelência. Isso sugere que a privação de sono, embora não seja o cenário ideal, geralmente está associada a horas extras de estudo e dedicação (o que cruza com nossos dados de alto estresse e altas notas).
* **A Inércia do Sono Excessivo:** Dormir mais de 8 horas apresenta a menor conversão para notas de elite (apenas 12.5%), indicando que o excesso de descanso pode estar associado a uma rotina menos intensa de estudos ou menor cobrança pessoal.

**Conclusão:** O desempenho acadêmico recompensa o equilíbrio. Sacrificar algumas horas de sono para estudar gera mais resultados do que dormir em excesso, mas a performance atinge seu limite máximo quando o aluno consegue manter 6 a 7 horas de descanso.

```python
grafico5_pct = df.groupby("screen_time_non_study")["cgpa_category"].value_counts(normalize=True).reset_index(name="Percentual")

df_grafico5 = px.bar(
    grafico5_pct,
    title="Tempo fora dos estudos",
    barmode="group",
    text_auto = ".1%",
    color="cgpa_category",
    x="screen_time_non_study",
    y="Percentual"

)

df_grafico5.show()
```

### O Impacto do Tempo de Tela Fora dos Estudos

Ao cruzar o tempo gasto em telas (lazer, redes sociais) com as notas finais, os dados quebram aquela regra de "quanto menos tela, melhor", revelando nuances sobre a qualidade do descanso:

* **O Excesso é Prejudicial:** Como esperado, o extremo superior (Mais de 6 horas) apresenta a menor taxa de notas de elite (13.9%), confirmando que o uso abusivo de telas concorre diretamente com o tempo e a qualidade do estudo.
* **O Paradoxo do Lazer Moderado:** Surpreendentemente, alunos que passam de 4 a 6 horas nas telas têm proporções de notas excelentes (16.7%) maiores do que aqueles que passam apenas 2 a 4 horas (16.1%).
* **O Grupo de Risco:** O dado mais curioso é que o grupo de "2 a 4 horas" concentra a maior taxa de alunos com notas baixas na faixa de 5.0 - 6.9 (24.5%). Isso sugere que não é apenas o *tempo* de tela que importa, mas como ele é usado: 4 a 6 horas podem representar um lazer estruturado (filmes, jogos) que alivia o estresse, enquanto 2 a 4 horas podem ser apenas distrações fragmentadas e ansiosas durante a rotina.

**Conclusão:** O lazer digital não é o grande vilão das notas, desde que não ultrapasse a marca de 6 horas. Um tempo de tela moderado (4-6 horas) parece fazer parte de uma rotina equilibrada e sustentável.

```python
grafico6_pct = df.groupby("study_consistency")["cgpa_category"].value_counts(normalize=True).reset_index(name="Percentual")

df_grafico6 = px.bar(
    grafico6_pct,
    title="Consistência",
    barmode="group",
    text_auto=".1%",
    color="cgpa_category",
    x="study_consistency",
    y="Percentual"
)

df_grafico6.show()
```

### O Poder da Consistência nos Estudos

Nossa última análise cruza a consistência da rotina de estudos com o desempenho final, provando que a disciplina diária supera o esforço esporádico:

* **O Prêmio da Constância:** O grupo "Mostly consistent" (Quase sempre consistente) é o líder absoluto no topo do boletim. 22.0% deles atingem a nota máxima (9.5 - 10.0), e formam o grupo com a menor taxa de notas baixas (apenas 10.9%).
* **O Custo da Inconstância:** Alunos que estudam de forma rara ("Rarely") sofrem o maior impacto negativo. Impressionantes 32.9% desse grupo tem as piores notas da turma (5.0 - 6.9), e eles praticamente não chegam ao topo (apenas 4.5% conseguem nota máxima).
* **O Meio-Termo:** O grupo "Sometimes" (Às vezes) fica exatamente no meio do caminho, mostrando um desempenho morno e distribuído.

**Conclusão:** A consistência é o pilar mais forte do desempenho acadêmico. O hábito regular de estudos protege o aluno da reprovação e é o caminho mais seguro para a excelência.
